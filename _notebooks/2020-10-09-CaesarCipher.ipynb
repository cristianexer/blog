```
def caesar(text,shift=3):
    def to_cipher(x,shift):
        x = ord(x)
        if x <= 122 and x >= 93:
            if (x + shift) >= 122:
                return chr(x + shift - 122 + 93)
            else:
                return chr(x + shift)
        if x <= 90 and x >= 65:
            if (x + shift) > 90:
                return chr(x + shift - 90 + 65)
            else:
                return chr(x + shift)
        
        return str(x + shift)

    return ''.join([ to_cipher(x,shift) for x in list(text)])
```
